Calcule o IMC do usu√°rio.


```python
weight_user = float(input('Please type your weight in Kg: '))

height_user = float(input('Please informe your height in meters: '))

IMC_user = weight_user / (height_user**2)

if IMC_user < 18.5:
    print('You are underweight.')
elif 18.5 <= IMC_user < 24.9:
    print('You are fit. Well done!')
elif 25 <= IMC_user <29.9:
    print('You are overweight.')
else:
    print('You should go to the gym.')

print(f'Your IMC is {IMC_user}')

```
