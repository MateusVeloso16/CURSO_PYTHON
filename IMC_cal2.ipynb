```python
def Calc_IMC(weight_user1, height_user2):

    IMC_user = weight_user1 / (height_user2**2)

    if IMC_user < 18.5:
        final_result = 'You are underweight.'
    elif 18.5 <= IMC_user < 24.9:
        final_result = 'You are fit. Well done!'
    elif 25 <= IMC_user <29.9:
        final_result = 'You are overweight.'
    else:
        final_result = 'You should go to the gym.'
    
    return final_result, IMC_user


weight_user = float(input('Please type your weight in Kg: '))

height_user = float(input('Please informe your height in meters: '))

Resultado = Calc_IMC(weight_user, height_user)

print(f'Your IMC e de {Resultado[1]:.2f}. {Resultado[0]}')




```
