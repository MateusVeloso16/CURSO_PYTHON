```python
def calculate(num_1, num_2):
    result_1 = num_1 + num_2
    result_2 = num_1 - num_2
    result_3 = num_1 * num_2
    result_4 = num_1 / num_2
    return result_1, result_2, result_3, result_4


first_number = int(input('Please inform the first number to calculate: '))
second_number = int(input('Please informe the second number to calculate: '))
result = calculate(first_number, second_number)
print(f'O resultado da soma dos dois numeros foi de {result[0]}.'
      f' \nO resultado da subtracao dos dois numeros foi de {result[1]}.'
      f' \nO resultado da multiplicacao dos dois numeros foi de {result[2]}.'
      f' \nO resultado da divisao dos dois numeros foi de {result[3]}')
```
